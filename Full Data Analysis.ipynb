```python
import string
import pickle
import pandas as pd
import numpy as np
import seaborn
import random
from matplotlib import pyplot as plt
%matplotlib inline


from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score

import sklearn
import sklearn.datasets
import sklearn.cross_validation
import sklearn.decomposition
import sklearn.grid_search
import sklearn.neighbors
import sklearn.metrics
```

## 4 Complete Data Analysis

#### 4.1 Dataframe with Full Features

```python
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
with open('data1000summarizedfull.txt', 'rb') as handle:
  trainingData = pickle.loads(handle.read())
```

```python
# make this into a function makeDataFrame
def makeDataFrame(index, trainingData):
    master_data = []
    for i in xrange(len(trainingData)):
        master_data.append(pd.DataFrame(dict(zip(index, trainingData[i])), columns = index))
    return master_data

# column names
index = ['games', 'KD', 'KDA', 'champKills', 'champDeaths', 'csPerGame', 'winRate', 'dmgPerGame', 'maxDeaths', 'maxKills', 
         'turretsPerGame', 'damageTakenPerGame', 'doublesPerGame', 'triplesPerGame', 'quadrasPerGame', 'pentasPerGame', 
         'firstBloodPerGame', 'goldPerGame', 'win']
nfeat = len(index)

training_df = pd.concat(makeDataFrame(index, trainingData))
training_df.head()
```

```python
# fix the KDA (since realisitically the most kills you can get is 30)
def fixDataFrame(df, cname, MAX_KILLS, MAX_ASSISTS):
    df[cname[0]] = [min(MAX_KILLS, x) for x in df[cname[1]]]
    df[cname[1]] = [min(MAX_KILLS + MAX_ASSISTS, x) for x in df[cname[1]]]
    
    # subset only games greater than 5 to avoid outliers
    df = df[df[cname[2]] > 2]

    # eliminate supports and jungler role (will expand on this)
    df = df[df[cname[3]] > 60]
    return df

full_df = fixDataFrame(training_df, ['KDA','KD','games','csPerGame'], 10, 10)
sample = np.sort(random.sample(xrange(full_df.shape[0]), 500))
full_df = full_df.iloc[sample]

full_df.head()
```

```python
Y = [int(x) for x in full_df['win']]
Y = np.asarray(Y)
X = full_df.drop('win', 1)
```

#### 4.2 Making a Random Forest Predictor

```python
# get training and test sets
X_train, X_test, Y_train, Y_test = sklearn.cross_validation.train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
scores_l = []
for i in xrange(40):
    forest = RandomForestClassifier(n_estimators = i+1) 
    scores_l.append(cross_val_score(forest, X_train, Y_train, cv=15))

seaborn.boxplot(scores_l, names = range(1,41))
plt.xlabel("Number of Trees")
plt.ylabel("Proportion Correctly Fitted")
plt.title("Cross Validation Score vs. Number of Trees in Random Forest")
plt.axhline(0.5, color = 'r')
plt.show()
```

```python
# prediction on test set (We randomly 20 trees in the forest 
# because it is efficient and the marginal increase in prediction doesn't improve)
forest = RandomForestClassifier(n_estimators = 20) 
forest.fit(X_train,Y_train)
pred = forest.predict(X_test)
pred_prob = forest.predict_proba(X_test)

print "The cross validation of the predictor is " + str(100*np.mean(pred == Y_test)) + "%."
```

```python
# prediction on test set
def cross_valid(classifier, parameter, X, Y):
    cross_validator = sklearn.grid_search.GridSearchCV(classifier, parameter, cv = 10)
    cross_validator.fit(X,Y)
    return cross_validator

parameter_best = cross_valid(RandomForestClassifier(random_state=42), {'n_estimators' : range(1,41)}, X_train, Y_train)
a = parameter_best.grid_scores_
```

```python
cross_val_estimate_accuracy = [a[x][1] for x in xrange(len(a))]
print "The accuracy of the predictor is "+str(max(cross_val_estimate_accuracy)) +" for the optimal number of trees in the forest."
```

```python
# We find the most important features. 
forest = RandomForestClassifier(n_estimators = 15) 
forest.fit(X,Y)
Z = forest.feature_importances_

seaborn.barplot(X.columns.values, Z)
plt.xticks(rotation = 90)
plt.title("Relative Importance of Features for Training Random Forest of 15 Trees")
plt.ylabel("Relative Ratio of Importance")
plt.show()
```

The most important feature is clearly the win rate. The other features seem equally important. The first blood per game has almost no influence. This makes sense since for each game there is only 1 first blood so the statistics is very small, unlike the other stats such as creep score where players normally have a minimum of 150 creep score or even double kills where there are usually 10 double kills in averaged length game. Since the other factors are almost equally important, we are going to select win rate and kill-death ratio to match the analysis of the previous section 3 for consistency. 

```python
# We select the two factors that matter the most. We select KD here to be consistent with our first dataset. 
X_sub = X[['winRate','KD']]
Xs_train, Xs_test, Y_train, Y_test = sklearn.cross_validation.train_test_split(X_sub, Y, test_size=0.33, random_state=42)
forest = RandomForestClassifier(n_estimators = 15) 
forest.fit(Xs_train,Y_train)
predSub = forest.predict(Xs_test)

print "The cross validation of the predictor is " + str(100*np.mean(predSub == Y_test)) + "%."
```

#### 4.3 Decision Surface Visuals

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.linear_model
import sklearn.svm

def plot_decision_surface(clf, X_train, Y_train, title):
    plot_step=0.1
    
    if X_train.shape[1] != 2:
        raise ValueError("X_train should have exactly 2 columnns!")
    
    x_min, x_max = X_train[:, 0].min() - plot_step, X_train[:, 0].max() + plot_step
    y_min, y_max = X_train[:, 1].min() - plot_step, X_train[:, 1].max() + plot_step
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    clf.fit(X_train,Y_train)
    if hasattr(clf, 'predict_proba'):
        Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
    else:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])    
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Reds)
    plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=plt.cm.Paired)
    plt.title(title)
    plt.xlabel("win rate")
    plt.ylabel("kd")
    plt.scatter([], [], color = 'grey', marker = 'o', label = 'loss')
    plt.scatter([], [], color = 'crimson', marker = 'o', label = 'win')
    plt.legend(bbox_to_anchor=(1.12, 1), loc=1, borderaxespad=0.)
    plt.show()
    
full_sub = full_df[["winRate", "KD"]]
X_sub = full_sub.as_matrix()
dTree = DecisionTreeClassifier()
dTree.fit(X_sub, Y)
plot_decision_surface(dTree, X_sub, Y, "Decision Tree Decision Surface")

forest_sub = RandomForestClassifier(n_estimators = 15) 
forest_sub.fit(X_sub, Y)
plot_decision_surface(forest_sub, X_sub, Y, "Random Forest Decision Surface")

svm = sklearn.svm.SVC(C = 100, gamma = 1.0, probability = True)
svm.fit(X_sub, Y)
plot_decision_surface(svm, X_sub, Y, "SVM Decision Surface")
```

These predictions are slightly better than the data with fewer features. Having more features improves the cross validation accuracy slightly. However we didn't expect a much better prediction since a lot of these features are correlated with the other features. For instance, first blood which is the statistic of whether a player gets the first kills during a game is clearly correlated with factors like kill death ratio.


Nevertheless, the random forest decision surface is very similar to the model for the data with fewer features. Surprisingly for both of them, a slightly lower than average historic Kill-Death ratio and a higher historic win rate are the best features for winning a game. In terms of the game, a self-less team oriented player is one who is likely to succeed. The shatters the preconception that one has to singly "carry" through getting kills and money. 

#### 4.4 F1 Scores

```python
scores_f = []
for i in xrange(40):
    forest = RandomForestClassifier(n_estimators = i+1) 
    scores_f.append(cross_val_score(forest, X_train, Y_train, cv=10, scoring = 'f1'))

seaborn.boxplot(scores_f, names = range(1,41))
plt.xlabel("Number of Trees")
plt.ylabel("F1 Score")
plt.title("F1 Score for Random Forest vs. Number of Trees")
plt.show()
```

This graph shows that we are in bounds of being both relatively precise and accurate. The default threshhold of 0.5 is pretty good as expected since exactly 50% of games are wins. Therefore, we do not feel it is necessary to alter the threshold value. This visual also confirms the random forest decision surface graphs (both of which show an acceptable precision and recall values). 

#### 4.5 Final Words

A prediction accuracy of around 62% on the test set is not too impressive. However, for a game as random League of Legends and for data that is hardly distinguishable, we are still satisfied with how the decision forest performs. 

The random forest predictor seems to be the best for categorical data described by multiple features. A KNN or SVM will not work due to the overlap of our data. We do not expect a high precision or recall. Nevertheless, our predictor allows us to make better predictions better than just randomly guessing. Dedicated gamers interested in improving their gameplay as well as casual player looking to "climb the ladder" would be interested in this data.
