```python
import string
import pickle
import pandas as pd
import numpy as np
import seaborn
from matplotlib import pyplot as plt
%matplotlib inline
import random


from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score

import sklearn
import sklearn.datasets
import sklearn.cross_validation
import sklearn.decomposition
import sklearn.grid_search
import sklearn.neighbors
import sklearn.metrics
```

## 3 Making a Predictor

#### 3.1 Tuning the Dataframe

```python
with open('data1000summarized.txt', 'rb') as handle:
  trainingData = pickle.loads(handle.read())
```

```python
# make this into a function makeDataFrame
def makeDataFrame(index, trainingData):
    master_data = []
    for i in xrange(len(trainingData)):
        master_data.append(pd.DataFrame(dict(zip(index, trainingData[i])), columns = index))
    return master_data

# column names
index = ['games', 'kd', 'kda', 'creep slain', 'win rate', 'damage dealt', 'win']
nfeat = len(index)

training_df = pd.concat(makeDataFrame(index, trainingData))
training_df.head()
```

```python
# fix the KDA (since realisitically the most kills you can get is 30)
def fixDataFrame(df, cname, MAX_KILLS, MAX_ASSISTS):
    df[cname[0]] = [min(MAX_KILLS, x) for x in df[cname[1]]]
    df[cname[1]] = [min(MAX_KILLS + MAX_ASSISTS, x) for x in df[cname[1]]]
    
    # subset only games greater than 5 to avoid outliers
    df = df[df[cname[2]] > 2]
    df = df[df[cname[2]] < 1000]
    
    # eliminate supports and jungler role (will expand on this)
    df = df[df[cname[3]] > 60]
    return df

carry_df = fixDataFrame(training_df, ['kda','kd','games','creep slain'], 20, 20)
sample = np.sort(random.sample(xrange(carry_df.shape[0]), 200))
carry_df = carry_df.iloc[sample]
'''
# scale data to [0,1]
MAX_DAMAGE = max(carry_df['damage dealt'])
MAX_CREEP = max(carry_df['creep slain'])
carry_df['damage dealt'] = carry_df['damage dealt']/MAX_DAMAGE
carry_df['creep slain'] = carry_df['creep slain']/MAX_CREEP
carry_df['win rate'] = carry_df['win rate']/100
'''
carry_df.head()
```

```python
Y = [int(x) for x in carry_df['win']]
Y = np.asarray(Y)
X = carry_df.drop('win', 1)
```

#### Visualizations

```python
def makeBoxPlot(df, x_feat, y_feat, names, y_label="", title=""):
    seaborn.boxplot([df[df[x_feat] == False][y_feat], carry_df[carry_df[x_feat] == True][y_feat]], names = names)
    plt.xlabel("win status")
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()
```

```python
for i in xrange(nfeat-1):
    makeBoxPlot(carry_df, index[nfeat-1], index[i], names = ['No','Yes'], 
            y_label = index[i], title = string.capitalize(index[i]) + ' of Winning and Losing Player')
```

Kd, kda, and win rate look slightly promising as factors to explain the differences. Games played is surprisingly an interesting factor. This suggests that people tend to win games better if they play a champion that they played frequently due to more experience. 

#### 3.2 Correlation of Features

#### 3.3 Building the Predictor

```python
# get training and test sets
X_train, X_test, Y_train, Y_test = sklearn.cross_validation.train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
plt.rcParams['figure.figsize'] = (10.0, 8.0)
# Cross Validation Scores for Different Number of Trees
scores_l = []
for i in xrange(40):
    forest = RandomForestClassifier(n_estimators = i+1) 
    scores_l.append(cross_val_score(forest, X_train, Y_train, cv=10))

seaborn.boxplot(scores_l, names = range(1,41))
plt.xlabel("Number of Trees")
plt.ylabel("Proportion Correctly Fitted")
plt.title("Cross Validation Score vs. Number of Trees in Random Forest")
plt.show()
```

#### Assessment of Model

```python
# prediction on test set (We randomly select 12 trees in the forest 
# because it is efficient and the marginal increase in prediction doesn't improve)
forest = RandomForestClassifier(n_estimators = 12) 
forest.fit(X_train,Y_train)
pred = forest.predict(X_test)
pred_prob = forest.predict_proba(X_test)

print "The cross validation of the predictor is " + str(100*np.mean(pred == Y_test)) + "%."
```

Here we find the two factors that best explain the predictions of the random forest model. We choose 15 trees since it is computationally effective and yields the best accuracy. The accuracy of the predictor hardly increases with more forests due to the marginally decreasing benefits of additional decision trees. The two most important factors are win rate and kill-death with win rate being much more effective. 

```python
plt.rcParams['figure.figsize'] = (8.0, 6.0)

forest = RandomForestClassifier(n_estimators = 15) 
forest.fit(X_train,Y_train)
Z = forest.feature_importances_

seaborn.barplot(X.columns.values, Z)
plt.xticks(rotation = 45)
plt.title("Relative Importance of Features for Training Random Forest of 15 Trees")
plt.ylabel("Relative Ratio of Importance")
plt.xlabel("Features")
plt.show()
```

A prediction accuracy of 0.60 is not very convincing so we do the svd and see if there is a hidden pattern. 

```python
svd = sklearn.decomposition.TruncatedSVD(n_components=2, random_state=42)
X_pca = svd.fit_transform(X_train)

plt.scatter(X_pca[:,0], X_pca[:,1], c = Y_train, cmap = plt.cm.Set1)
plt.xlabel("first PC")
plt.ylabel("second PC")
plt.title("First 2 PCA of Digits Data Set")
plt.scatter([], [], color = 'crimson', marker = 'o', label = 'win')
plt.scatter([], [], color = 'grey', marker = 'o', label = 'loss')
plt.legend(loc = 1)
plt.show()
```

There doesn't seem to be an easy way to distinguish the two groups from each other so a method such as SVM or KNN would not be viable. These require us to be able to distinguish the location of the data points. SVM draws a boundary and KNN makes decisions based on the neighbors of the points. But since all the points are close together, it's hard to use these algorithms.

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.linear_model
import sklearn.svm

def plot_decision_surface(clf, X_train, Y_train, title):
    plot_step=0.1
    
    if X_train.shape[1] != 2:
        raise ValueError("X_train should have exactly 2 columnns!")
    
    x_min, x_max = X_train[:, 0].min() - plot_step, X_train[:, 0].max() + plot_step
    y_min, y_max = X_train[:, 1].min() - plot_step, X_train[:, 1].max() + plot_step
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    clf.fit(X_train,Y_train)
    if hasattr(clf, 'predict_proba'):
        Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
    else:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])    
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Reds)
    plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=plt.cm.Paired)
    plt.title(title)
    plt.xlabel("win rate")
    plt.ylabel("kd")
    plt.scatter([], [], color = 'crimson', marker = 'o', label = 'win')
    plt.scatter([], [], color = 'grey', marker = 'o', label = 'loss')
    plt.legend(bbox_to_anchor=(1.12, 1), loc=1, borderaxespad=0.)
    plt.show()
    
carry_sub = carry_df[["win rate", "kd"]]
X_sub = carry_sub.as_matrix()
dTree = DecisionTreeClassifier()
dTree.fit(X_sub, Y)
plot_decision_surface(dTree, X_sub, Y, "Decision Tree Decision Surface")

forest_sub = RandomForestClassifier(n_estimators = 15) 
forest_sub.fit(X_sub, Y)
plot_decision_surface(forest_sub, X_sub, Y, "Random Forest Decision Surface")

svm = sklearn.svm.SVC(C = 100, gamma = 1.0, probability = True)
svm.fit(X_sub, Y)
plot_decision_surface(svm, X_sub, Y, "SVM Decision Surface")
```

The SVM is not a good predictor. This is because the SVM draws a boundary and there's no distinguishable boundary between the win and loss points. However the Random Forest Decision surface seems to make some intutitive sense. It predicts that players who enter the game with a higher win rate on the champion they are playing are more likely to win. However an interesting pattern to note is that the decision surface is bottom-right. It predicts that players with a lower kda on their champion are more likely to win. This is somewhat contradictory as one would suppose that a player with more kills and less death would tends to win since we believe that kills are positively correlated with chances of winning. 

Our data tells another story: it suggests that the game League of Legends a relatively lower kill death ratio improves the chance of winning. Our data tells us that those players who aren't greedy for kills have a better chance of winning, which makes sense since league is a team game that requires cooperation among the 5 members. Our data also tells us that the popularly believed concept that a skilled individual can carry a team is incorrect, which is represented by the cluster of data points with high kda and high winrate which are mostly losses. A team where all players are equally skills, represented by the cluster of wins with moderately high kda and high past win rates, is the most successful.
