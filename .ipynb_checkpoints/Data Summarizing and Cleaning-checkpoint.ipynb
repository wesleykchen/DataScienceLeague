```python
# below are all functions from Data Grabbing ipython notebook as I wasn't sure how to import though I know it's possible

def createChampionLookup():
    # create map of chapmion name to ID and vice versa
    champion2id = {}
    id2champion = {}
    
    url = 'https://na.api.pvp.net/api/lol/na/v1.2/champion?api_key=' + apiKey
    query = requests.get(url).json()
    
    for champ in query['champions']:
        champId = str(champ['id'])
        url2 = 'https://na.api.pvp.net/api/lol/static-data/na/v1.2/champion/'+champId+'?champData=info&api_key=' + apiKey
        query2 = requests.get(url2).json()
        champName = query2['name']
        champion2id[champName.lower()] = champId
        id2champion[champId] = champName.lower()
    return champion2id, id2champion

champion2id, id2champion = createChampionLookup()
```

```python
# ids are list of ints, not strings

def getStats(summonerIds, championIds):
    
    # interesting stats to look for
    games = []
    KD = []
    KDA = []
    champKills = []
    champDeaths = []
    # this is limited by data availability already precompiled by Riot's API
    csPerGame = []
    winRate = []
    dmgPerGame = []
    # below this are the new, full ratios
    maxDeaths = []
    maxKills = []
    turretsPerGame = []
    damageTakenPerGame = []
    doublesPerGame = []
    triplesPerGame = []
    quadrasPerGame = [] 
    pentasPerGame = []
    firstBloodPerGame = []
    goldPerGame = []
    
    

    # get champion statistics per summoner
    for idx, summId in enumerate(summonerIds):
        url = 'https://na.api.pvp.net/api/lol/na/v1.3/stats/by-summoner/'+str(summId)+'/ranked?season=SEASON4&api_key='+apiKey
        query = requests.get(url)
        
        # ensure we don't go over the API rate
        while query.status_code != 200:
            time.sleep(1)
            query = requests.get(url)

        data = query.json()
        foundFlag = False
        for champ in data['champions']:
            if champ['id'] == championIds[idx]:
                # some syntax shortcuts for readibility
                stats = champ['stats']
                numGames = float(stats['totalSessionsPlayed'])
                # TODO fix this temporary test
                if numGames == 0:
                    numGames = 1
                killsPerGame = float(stats['totalChampionKills'])/numGames
                assistsPerGame = float(stats['totalAssists'])/numGames
                deathsPerGame = float(stats['totalDeathsPerSession'])/numGames
                # give perfect games high scores - though not infinity
                # should not affect that badly
                if deathsPerGame == 0:
                    deathsPerGame = 0.001
                cs = float(stats['totalMinionKills'])
                wins = float(stats['totalSessionsWon'])
                damage = float(stats['totalDamageDealt'])
                maxDeathsVal = float(stats['maxChampionsKilled'])
                maxKillsVal = float(stats['totalMinionKills'])
                turrets = float(stats['totalMinionKills'])
                damageTaken = float(stats['totalDamageTaken'])
                doubles = float(stats['totalDoubleKills'])
                triples = float(stats['totalTripleKills'])
                quadras = float(stats['totalQuadraKills'])
                pentas = float(stats['totalPentaKills'])
                firstBloods = float(stats['totalFirstBlood'])
                gold = float(stats['totalGoldEarned'])

                # add per game numbers to lists
                games.append(numGames)
                KD.append(killsPerGame/deathsPerGame)
                KDA.append((killsPerGame+assistsPerGame)/deathsPerGame)
                champKills.append(killsPerGame)
                champDeaths.append(deathsPerGame)
                csPerGame.append(cs/numGames)
                winRate.append(wins/numGames*100)
                dmgPerGame.append(damage/numGames)
                maxDeaths.append(maxDeathsVal)
                maxKills.append(maxKillsVal)
                turretsPerGame.append(turrets/numGames)
                damageTakenPerGame.append(damageTaken/numGames)
                doublesPerGame.append(doubles/numGames)
                triplesPerGame.append(triples/numGames)
                quadrasPerGame.append(quadras/numGames)
                pentasPerGame.append(pentas/numGames)
                firstBloodPerGame.append(firstBloods/numGames)
                goldPerGame.append(gold/numGames)
                foundFlag = True

            if champ == data['champions'][-1] and not(foundFlag):
                games.append(0.0)
                KD.append(0.0)
                KDA.append(0.0)
                champKills.append(0.0)
                champDeaths.append(0.0)
                csPerGame.append(0.0)
                winRate.append(0.0)
                dmgPerGame.append(0.0)
                maxDeaths.append(0.0)
                maxKills.append(0.0)
                turretsPerGame.append(0.0)
                damageTakenPerGame.append(0.0)
                doublesPerGame.append(0.0)
                triplesPerGame.append(0.0)
                quadrasPerGame.append(0.0)
                pentasPerGame.append(0.0)
                firstBloodPerGame.append(0.0)
                goldPerGame.append(0.0)

    return games, KD, KDA, champKills, champDeaths, csPerGame, winRate, dmgPerGame, maxDeaths, maxKills, turretsPerGame, damageTakenPerGame, doublesPerGame, triplesPerGame, quadrasPerGame, pentasPerGame, firstBloodPerGame, goldPerGame
```

```python
datafile = open('data5.txt', 'r')
gamedata = json.loads(datafile.read())

# did not check for the team that tehy are on yet - assuming API has it 5 then 5 - but can confirm later

# DICTIONARY INDEXED BY GAME# (from 0 to max) with 6 lists each, 
allGamesStats = {}

counter = 0
for game in gamedata:
    teamMap = {}
    championMap = {}
    resultMap = {}
    
    # default one winning team
    winningTeamId = 100
    
    for team in game['teams']:
        if (team['winner']):
            winningTeamId = team['teamId']
            
    # get information of participants in game as dictionary to perserve relative order
    
    for player in game['participantIdentities']:
        teamMap[player['participantId']] = player['player']['summonerId']
    
    for champion in game['participants']:
        championMap[champion['participantId']] = champion['championId']
        resultMap[champion['participantId']] = champion['teamId']
        
    
    # convert dictionary to list in respective order between champion and summoner
    teamList = []
    championList = []
    resultList = []
    for x in xrange(1,11):
        teamList.append(teamMap[x])
        championList.append(championMap[x])
        if (resultMap[x] == winningTeamId):
            resultList.append(True)
        else:
            resultList.append(False)

    allGamesStats[counter] = getStats(teamList, championList)
    allGamesStats[counter] = allGamesStats[counter] + (resultList,)
    counter += 1
    print counter
    
print allGamesStats[0]
```

```python
with open('data1000summarizedfull.txt', 'wb') as handle:
  pickle.dump(allGamesStats, handle)
```
